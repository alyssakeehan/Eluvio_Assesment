### Feature processing in pandas

1. Converting a string type categorical value into one-hot encoded features

https://medium.com/@michaeldelsole/what-is-one-hot-encoding-and-how-to-do-it-f0ae272f1179

```python
import pandas as pd
education_one_hot = pd.get_dummies(df['education'])
education_one_hot
```

We can also convert it to ordinal data, but it's usually less preferable than onehot encoding.

```python
education_ordinal, class_index = pd.factorize(df['education'])
print(education_ordinal)
print(class_index)
```
